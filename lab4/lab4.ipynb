```python
import pandas as pd
import numpy as np

# --- КРОК 1: ЗАВАНТАЖЕННЯ ТА ПІДГОТОВКА ДАНИХ ---
url = "https://raw.githubusercontent.com/selva86/datasets/master/Cars93_miss.csv"
df = pd.read_csv(url)
df['Price'] = df['Price'].fillna(df['Price'].mean())
df['Horsepower'] = df['Horsepower'].fillna(df['Horsepower'].mean())
df['Weight'] = df['Weight'].fillna(df['Weight'].mean())
print("--- КРОК 1 ЗАВЕРШЕНО: ДАНІ ЗАВАНТАЖЕНО ТА ОЧИЩЕНО ---")

# --- КРОК 2: СТВОРЕННЯ НОВИХ ОЗНАК ---
df['Power_Weight_Ratio'] = df['Horsepower'] / df['Weight']
df['Price_Level'] = pd.qcut(df['Price'], q=3, labels=['Budget', 'Standard', 'Premium'])
print("\n--- КРОК 2 ЗАВЕРШЕНО: НОВІ КОЛОНКИ ДОДАНО ---")
print(df[['Manufacturer', 'Model', 'Power_Weight_Ratio', 'Price_Level']].head())

# --- КРОК 3: ГРУПУВАННЯ ТА АГРЕГАЦІЯ ---
analysis_by_type = df.groupby('Type').agg({
    'Price': ['mean', 'min', 'max'],
    'MPG.city': 'mean',
    'Horsepower': 'mean'
})
print("\n--- КРОК 3 ЗАВЕРШЕНО: ГРУПУВАННЯ ВИКОНАНО ---")
print(analysis_by_type)

# --- КРОК 4: ЗВЕДЕНІ ТАБЛИЦІ ---
pivot_table = df.pivot_table(index='Origin', columns='Price_Level', values='Model', aggfunc='count', fill_value=0)
print("\n--- КРОК 4 ЗАВЕРШЕНО: ПІВОТ СТВОРЕНО ---")
print(pivot_table)

# --- КРОК 5: СКЛАДНА ФІЛЬТРАЦІЯ ДАНИХ ---
print("\n--- КРОК 5: РЕЗУЛЬТАТ СКЛАДНОЇ ФІЛЬТРАЦІЇ (ОПТИМАЛЬНІ АВТО) ---")
efficient_and_powerful = df[(df['Horsepower'] > df['Horsepower'].mean()) & (df['MPG.city'] > df['MPG.city'].mean())]
# Виводимо тільки перші 10 знайдених, щоб VS Code не обрізав вивід
print(efficient_and_powerful[['Manufacturer', 'Model', 'Price', 'Horsepower', 'MPG.city']].head(10))

print("\n--- ЛАБОРАТОРНУ РОБОТУ №4 ЗАВЕРШЕНО УСПІШНО ---")
```
